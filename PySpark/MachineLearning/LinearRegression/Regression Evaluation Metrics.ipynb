# Linear Regression Evaluation Metrics

## Mean Absolute Error
- Mean of the absolute value of the errors, just average error
- Mean of True value - Predicted value

```python
from IPython.display import Image
Image(filename='Mean Absolute Error.png') 
```

## Mean Squared Error
- Mean of the squared errors
- Larger errors are noted more than with MAE, making MSE more popular
- Larger errors more visible
- different units than original y, now squared

```python
from IPython.display import Image
Image(filename='Mean Squared Error.png') 
```

## Root Mean Square Error
- Root of the mean of the squared errors
- Most popular (has same units as y)

```python
from IPython.display import Image
Image(filename='Root Mean Absolute Error.png') 
```

## R Squared Values
- coefficient of determination
- measuer of how much variance your model accounts for
- always between 0 and 1
- example: model explains 60 percent of the variance
