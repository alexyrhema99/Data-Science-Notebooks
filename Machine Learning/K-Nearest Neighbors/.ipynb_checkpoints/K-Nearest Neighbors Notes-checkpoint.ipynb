# K-Nearest Neighbors

### What are the points closest to the point in question's classification?
### Prediction Algorithm: <br> 1. Calculate the distancee from x to all points in your data <br> 2. Sort the points in your data by increasing distance from x. <br> 3. Predict the majority label of the "k" closest points.

### Pros: <br> Very simple <br> Training is trivial <br> Works with any number of classes <br> Easy to add more data <br> Few parameters: K, Distance Metric

### Cons: <br> High Prediction Cost <br> Not good with high dimensional data <br> Categorical Features don't work well
